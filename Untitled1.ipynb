```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = 'https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv'
df = pd.read_csv(url)
df.rename(columns={'Unnamed: 0':'model'}, inplace=True)
df = df.set_index('model')
 
# Prepare a vector of color mapped to the 'cyl' column
my_palette = dict(zip(df.cyl.unique(), ["orange","yellow","brown"]))
row_colors = df.cyl.map(my_palette)
 
# plot
sns.clustermap(df, metric="correlation", method="single", cmap="Blues", standard_scale=1, row_colors=row_colors)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
#url = 'https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv'
df1 = pd.read_csv("/Users/paritoshkumar/Desktop/saba/resequencing/heatmap/mtb_combined_percent_3.csv")
#df1.drop(df1.columns[len(df1.columns)-1], axis=1, inplace=True)
df1 = df1.set_index('id')
my_palette1 = dict(zip(df1.blosum.unique(), ["darkred","firebrick","salmon", "mistyrose", "palegreen", "greenyellow", "lawngreen"]))
row_colors1 = df1.blosum.map(my_palette1)
df1 = df1.drop("blosum", axis=1)
g= sns.clustermap(df1, metric="correlation", method="single", cmap="Blues", row_cluster=False, col_cluster=False, standard_scale=None, row_colors=row_colors1)
g.fig.subplots_adjust(right=0.7)
g.ax_cbar.set_position((0.9, .2, .03, .4))
handles = [Patch(facecolor=my_palette1[name]) for name in my_palette1]
plt.legend(handles, my_palette1, title='blosum',
           bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure, loc='upper right')
plt.show()
```
