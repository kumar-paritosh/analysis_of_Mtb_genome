```python
import pycircos
import matplotlib.pyplot as plt
Garc    = pycircos.Garc
Gcircle = pycircos.Gcircle
```

```python
#Set chromosomes
circle = Gcircle() 
with open("chr_input.txt") as f:
    f.readline()
    for line in f:
        print(line)
        line   = line.rstrip().split(",")
        name   = line[0]
        length = int(line[-1])
        print(length)
        #arc    = Garc(arc_id=name, size=length, interspace=0, raxis_range=(0,10),  label_visible=True)
        arc    = Garc(arc_id=name, size=length, interspace=0, raxis_range=(930,990),  
                      label="M. tuberculosis str Rv780", label_visible=False,
                     facecolor="#f0dae0")
        circle.add_garc(arc) 
```

```python
circle.set_garcs() 
```

```python
import collections
values_all   = [] 
arcdata_dict = collections.defaultdict(dict)
with open("genes.txt") as f:
    f.readline()
    for line1 in f:
        line1  = line1.rstrip().split(",")
        name  = line1[0]     
        start = int(line1[1])-1
        if int(line1[1]) > int(line1[2]):
            width = int(line1[1])-(int(line1[2])-1)
        else:
            width = int(line1[2])-(int(line1[1])-1) 
        if name not in arcdata_dict:
            arcdata_dict[name]["positions"] = []
            arcdata_dict[name]["widths"]    = [] 
            arcdata_dict[name]["colors"]    = [] 
        arcdata_dict[name]["positions"].append(start) 
        arcdata_dict[name]["widths"].append(4000)
        arcdata_dict[name]["colors"].append("#A0061D")
        
#print(arcdata_dict)
for key in arcdata_dict:
#    print(key)
    circle.barplot(key, data=[1]*len(arcdata_dict[key]["positions"]), positions=arcdata_dict[key]["positions"], 
                   width=arcdata_dict[key]["widths"], raxis_range=[930,990], facecolor=arcdata_dict[key]["colors"], spine=True)    
    
```

```python
#scatter plot
import re
values_all   = [] 
arcdata_dict = collections.defaultdict(dict)
with open("RvGP1.csv") as f:
    f.readline()
    for line in f:
        line  = line.rstrip().split(",")
        name  = line[0]     
        start = int(line[1])-1
        #end   = int(line[2]) 
        mid   = start
        if re.findall(r"A|T", line[2]) and re.findall(r"G|C", line[3]):
            value= float(0.5)
            color = "#ff00ff"
        elif re.findall(r"A|T", line[3]) and re.findall(r"G|C", line[2]):
            value= float(0.5)
            color = "#ff00ff"
        else:
            value= float(0.01)
            color = "#080808"
        #value = float(line[-1] ) 
        values_all.append(value) 
        if name not in arcdata_dict:
            arcdata_dict[name]["positions"] = []
            arcdata_dict[name]["values"] = []
            arcdata_dict[name]["colors"] = []
        arcdata_dict[name]["positions"].append(mid) 
        arcdata_dict[name]["values"].append(value)
        arcdata_dict[name]["colors"].append(color)
        
    
vmin, vmax = min(values_all), max(values_all) 
for key in arcdata_dict:
    circle.scatterplot(key, data=arcdata_dict[key]["values"], positions=arcdata_dict[key]["positions"], 
                       rlim=[0,1], raxis_range=(890,940), facecolor=arcdata_dict[key]["colors"], spine=False) 
```
